# EDA for TS115 Dataset

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(f"../Data/Preprocessed/ts115_preprocessed.csv")
df
```

```python
# Define the 8 DSSP classes
dssp8_classes = ['C', 'B', 'E', 'G', 'I', 'H', 'S', 'T']

# Flatten all predicted and actual secondary structure strings into lists of single characters
flatt = [ss for seq in df['dssp8'] for ss in seq]

# Count the occurrences of each class
counts = pd.Series(flatt).value_counts().reindex(dssp8_classes, fill_value=0)

# Print the counts
print("Counts of DSSP 8-Class Secondary Structure Predictions:")
print(counts)

# Plot the counts
counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of DSSP 8-Class Secondary Structure Predictions')
plt.xlabel('DSSP 8-Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
